# Execute queries with the original TREC COVID dataset

```python
!python3.8 {PYSERINI_TOOLS_FOLDER}convert_collection_to_jsonl.py \
    --collection-path {TREC_COVID_ORIGINAL_TITLE_TEXT_MERGED_FILENAME} \
    --output-folder {TREC_COVID_ORIGINAL_FOLDER}
```

```python
os.chdir(PYSERINI_FOLDER)
```

```python
os.environ["ANSERINI_CLASSPATH"]="/media/eduseiti/bigdata01/unicamp/ia368v_dd/anserini/target"
```

```python
!python3.8 -m pyserini.index.lucene \
    --collection JsonCollection \
    --input {WORK_FOLDER}{TREC_COVID_ORIGINAL_FOLDER} \
    --index {WORK_FOLDER}{TREC_COVID_ORIGINAL_INDEX_FOLDER} \
    --generator DefaultLuceneDocumentGenerator \
    --threads 9 \
    --storePositions --storeDocvectors --storeRaw
```

```python
execution_timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

pyserini_runfile = "run.trec_covid_original_{}.txt".format(execution_timestamp)
```

```python
!python3.8 -m pyserini.search.lucene \
    --index {WORK_FOLDER}{TREC_COVID_ORIGINAL_INDEX_FOLDER} \
    --topics {WORK_FOLDER}{TREC_COVID_QUERIES} \
    --output {WORK_FOLDER}{TREC_COVID_ORIGINAL_FOLDER}/runs/{pyserini_runfile} \
    --output-format trec \
    --hits 1000 \
    --bm25 --k1 0.82 --b 0.68 \
    --threads 8
```

```python
!tools/eval/trec_eval.9.0.4/trec_eval -c -mrecall.1000 -mmap -mndcg_cut.10 -mrecip_rank.100 \
    {WORK_FOLDER}{TREC_COVID_QRELS} {WORK_FOLDER}{TREC_COVID_ORIGINAL_FOLDER}/runs/{pyserini_runfile}
```
